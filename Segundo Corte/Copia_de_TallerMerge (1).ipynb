## 3. Integración de Datos mediante Joins

El objetivo es unir la información de las distintas tablas para poder responder a preguntas de negocio. En este ejercicio central, se utilizarán joins para obtener insights combinando datos de carreras, resultados y pilotos.

### 3.1 Unir `results_df` y `drivers_df`

Para analizar la participación y resultados de cada piloto, realizamos un merge entre `results_df` y `drivers_df` utilizando la clave `driverId`.

**Pregunta clave:** ¿Cuántos pilotos distintos han participado en la F1?

**_Respuesta:_** Han participado 857 pilotos

Utiliza el DataFrame resultante para explorar esta pregunta.

```python
df_resultados = pd.merge(results_df, drivers_df, on='driverId', how='inner')
df_resultados
```

```python
num_pilotos=df_resultados['driverId'].nunique()
num_pilotos
```

### 3.2. Unir `df_resultados` con `races_df`

Ahora, para relacionar los resultados con cada carrera, se efectúa la unión entre el DataFrame obtenido (`df_resultados`) y `races_df` a través de la clave `raceId`.

**Pregunta clave:** ¿Qué información adicional sobre cada carrera (fecha, circuito, etc.) se puede incorporar para enriquecer el análisis?

Explora y analiza cómo afecta esta integración en el conjunto de datos final.

**_Respuesta:_** Para enriquecer el análisis se puede usar información extra como la ubicación del circuito ya que permite tener una dimensión geográfica y explorar el rendimiento de los pilotos en los distintos tipos de pistas que existen en los países. Otra información que se puede agregar es el número de vueltas ya que se podría agrupar los resultados según el número de rondas para visualizar el progreso de los pilotos a lo largo de la temporada.

```python
df_completo = pd.merge(df_resultados, races_df, on='raceId', how='inner')
df_completo
```

```python
df_completo.info()
```

### Ejercicio 2: Piloto con Más Victorias

Filtra el DataFrame combinado para quedarte con aquellos registros donde la posición final (`position`) sea 1, lo que indica una victoria. Luego, agrupa por piloto para determinar quién ha ganado más carreras.

**Pregunta para los estudiantes:** _¿Qué piloto se destaca por su número de victorias y cómo se relaciona este dato con otros indicadores de rendimiento?_

_**Respuesta:**_ El piloto que se destaca por su número de victorias es Lewis Hamilton con 103 victorias. Este dato se relaciona con indicadores de rendimiento como la puntuación total acumulada y los podios, lo que lo coloca entre los mejores pilotos.  

```python
# Filtrar victorias (posición 1)
victorias_df = df_completo[df_completo['position'] == '1']
victorias_df
```

```python
pilotos_victorias = victorias_df.groupby('driverRef')['position'].count().sort_values(ascending=False)
pilotos_victorias
```

### Ejercicio 3: Piloto con Más Puntos

Agrupa el DataFrame combinado para sumar el total de puntos (`points`) obtenidos por cada piloto a lo largo de su carrera.

**Pregunta para los estudiantes:** _¿Existe correlación entre el número de victorias y la suma total de puntos? ¿Qué puede influir en esta relación?_

**_Respuesta:_** Sí existe una correlación entre el número de victorias y la suma total de puntos ya que entre más victorias haya, más puntos totales tendran los pilotos. Factores como el sistema de puntuación a lo largo de la historia, la consistencia del piloto en obtener puntos en cada carrera, las retiradas por problemas mecánicos o accidente y la duración de la carrera del piloto; pueden influir en esta relación.

```python
puntos_piloto = df_completo.groupby('driverRef')['points'].sum().sort_values(ascending=False)
puntos_piloto
```

```python
top_5_pilotos = puntos_piloto.head(5)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_5_pilotos.index, top_5_pilotos.values, color=["#F2EB88","#F29F05","#F28705","#D90404","#730202"])
plt.xlabel("Piloto")
plt.ylabel("Puntos Totales")
plt.title("Top 5 Pilotos con Más Puntos")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Ejercicio 4: Distribución de Nacionalidades de los Pilotos

Analiza la columna `nationality` presente en `drivers_df` para determinar qué nacionalidades están mejor representadas en la F1. Realiza un gráfico de pastel para visualizar la distribución de pilotos por país.

**Debate:** ¿Qué factores históricos o deportivos crees que puedan explicar esta distribución?

_**Respuesta:**_ Factores que podrían explicar esta distribución son: países con una larga tradición en el automovilismo tienden a tener más pilotos en la F1, la popularidad de la F1 en un país influye en el número de pilotos que produce, el écito de pilotos de una nacionalidad puede inspirar a otros

```python
nacionalidades_counts=df_completo['nationality'].value_counts()
nacionalidades_counts
```

```python
plt.figure(figsize=(10, 10))
plt.pie(nacionalidades_counts.values, labels=nacionalidades_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de los pilotos de F1 por nacionalidad')
```

```python
top_9 = nacionalidades_counts[:9]
other_count = nacionalidades_counts[9:].sum()
pie_data = pd.concat([top_9, pd.Series([other_count], index=['Others'])])
pie_data
```

```python
plt.figure(figsize=(10, 10))
plt.pie(pie_data.values, labels=pie_data.index, autopct='%1.1f%%', startangle=90, colors=["#CCD6DF","#8FB8C6","#D9674E","#C05048","#D9AB91","#E69D87","#F2E8DC"] * len(pie_data),textprops={"fontsize":14})
plt.title('Distribución de pilotos de F1 por nacionalidad (Top 9 + Otros))')
plt.show()
```

### Ejercicio 5: Evolución del Número de Carreras por Año

1. **Integración y Calidad de Datos:** ¿Qué inconvenientes pueden surgir al unir tablas que tienen registros incompletos o duplicados? ¿Cómo solucionarlo?

_**Respuesta:**_ Al unir tablas con registros incompletos se podrían obtener resultados sesgados y que se pierdan registros. Al registrar datos duplicados se podrían distorsionar los resultados ya que se contaran varias veces los mismos. Esto se podría solucionar utilizando un "Outter Join" para rellenar los valores que aparezcan en NaN.

2. **Estrategias de Join:** ¿En qué casos es más adecuado usar un `inner join` y cuándo un `outer join` para evitar perder información relevante?

_**Respuesta:**_ Inner Join es adecuado usarlo cuando se necesite información que esté presente en ambas tablas y solo se interese la intersección de los datos. Outter Join es adecuado usarlo cuando se quiera preservar toda la inofrmación de una o ambas tablas, incluso su no hay coincidencias, o cuando se quiera identificar registros que faltan en una de las tablas.
