```python
merged_games_leagues = pd.merge(games_df, leagues_df, on='leagueID', how='left')
merged_games_leagues
```

```python
games_df['total_goals'] = games_df['homeGoals'] + games_df['awayGoals']
goals_by_league = games_df.groupby('leagueID')['total_goals'].mean().reset_index()
goals_by_league = goals_by_league.merge(leagues_df, left_on='leagueID', right_on='leagueID')[['name', 'total_goals']]
goals_by_league.rename(columns={'name': 'Liga', 'total_goals': 'Goles por Partido'}, inplace=True)
print(goals_by_league)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="Liga", y="Goles por Partido", data=goals_by_league, palette="viridis")
plt.xlabel("Liga", fontsize=12)
plt.ylabel("Goles por Partido", fontsize=12)
plt.title("Promedio de Goles por Partido en Cada Liga", fontsize=14)
plt.xticks(rotation=45, ha="right", fontsize=10)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points')

plt.tight_layout()
plt.show()
```

**Evolución de goles por partido de quipo local y visitante**

```python
import pandas as pd
import matplotlib.pyplot as plt
games_df['total_goals'] = games_df['homeGoals'] + games_df['awayGoals']
goals_by_season = games_df.groupby('season')['total_goals'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(goals_by_season['season'], goals_by_season['total_goals'], marker='o')
plt.xlabel("Año")
plt.ylabel("Goles por Partido")
plt.title("Evolución de los Goles por Partido por Año")
plt.grid(True)
plt.show()
```

**Comparación de goles local vs visitante**

```python
home_goals_mean = games_df['homeGoals'].mean()
away_goals_mean = games_df['awayGoals'].mean()
plt.figure(figsize=(8, 6))
sns.barplot(x=['Local', 'Visitante'], y=[home_goals_mean, away_goals_mean], palette="Set2")
plt.xlabel("Tipo de Equipo")
plt.ylabel("Media de Goles")
plt.title("Comparación de Goles: Local vs Visitante")
plt.show()
```

**Equipos con ventaja o deventaja inusual**

```python
games_df['goal_diff'] = games_df['homeGoals'] - games_df['awayGoals']
home_teams = games_df.groupby('homeTeamID')['goal_diff'].agg(['mean', 'std'])
Q1 = home_teams['mean'].quantile(0.25)
Q3 = home_teams['mean'].quantile(0.75)
IQR = Q3 - Q1
home_outliers = home_teams[(home_teams['mean'] > Q3 + 1.5 * IQR)]

away_teams = games_df.groupby('awayTeamID')['goal_diff'].agg(['mean', 'std'])
Q1_away = away_teams['mean'].quantile(0.25)
Q3_away = away_teams['mean'].quantile(0.75)
IQR_away = Q3_away - Q1_away
away_outliers = away_teams[(away_teams['mean'] < Q1_away - 1.5 * IQR_away)]

print("Equipos con ventaja inusual en casa:")
print(home_outliers)
print("\nEquipos con desventaja inusual fuera de casa:")
print(away_outliers)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=home_teams['mean'])
plt.title('Distribución de la diferencia de goles para equipos locales')
plt.xlabel('Diferencia de goles (media)')
plt.show()

outliers_values = home_outliers['mean'].values
plt.scatter(outliers_values, [1] * len(outliers_values), color='red', marker='x', s=100, label='Outliers')
plt.legend()
plt.show()
```
